## 2. Теоретические сведения

### 2.1. Что такое LLM

**LLM (Large Language Model)** — большая языковая модель, обученная на огромных текстовых корпусах для понимания и генерации естественного языка.

**Ключевые особенности:**
- Обучаются на миллиардах токенов (слов/подслов)
- Используют архитектуру Transformer (attention mechanism)
- Способны к zero-shot и few-shot обучению
- Могут выполнять задачи: генерация, классификация, перевод, QA, суммаризация

### 2.2. Hugging Face Transformers

**Hugging Face** — платформа и библиотека для работы с предобученными моделями.

| Компонент | Назначение |
|-----------|------------|
| `pipeline()` | Быстрый запуск готовых задач |
| `AutoModel` | Загрузка любой модели |
| `AutoTokenizer` | Токенизация текста |
| `Hugging Face Hub` | Хранилище 500,000+ моделей |

### 2.3. Основные задачи NLP

| Задача | Описание | Пример |
|--------|----------|--------|
| `text-generation` | Генерация продолжения текста | Чат-бот, копирайтинг |
| `text-classification` | Классификация текста | Анализ тональности |
| `question-answering` | Ответ на вопрос по контексту | Поиск в документах |
| `summarization` | Сжатие текста | Краткий пересказ |
| `translation` | Перевод между языками | EN ↔ RU |

### 2.4. Токенизация

**Токенизатор** преобразует текст в последовательность чисел (токенов), которые понимает модель.

```python
Текст: "Привет, мир!"
Токены: [101, 7592, 2088, 999, 102]  # IDшки в словаре модели
```

### 4.2. Работа с pipeline()

**Pipeline** — самый простой способ использовать модели для стандартных задач.

````python
# ЗАДАЧА 1: Генерация текста
generator = pipeline("text-generation", model="Qwen/Qwen2.5-0.5B-Instruct", trust_remote_code=True)

result = generator("Искусственный интеллект — это", max_length=50, num_return_sequences=1)
print("Генерация текста:")
print(result[0]['generated_text'])
````

````python
# ЗАДАЧА 2: Классификация тональности
classifier = pipeline("sentiment-analysis")

texts = [
    "Этот товар превзошёл все мои ожидания! Очень доволен.",
    "Ужасное качество, деньги выброшены на ветер.",
    "В целом нормально, но есть куда расти."
]

print("\nКлассификация тональности:")
for text in texts:
    result = classifier(text)[0]
    print(f"\nТекст: {text}")
    print(f"  → {result['label']} (уверенность: {result['score']:.2%})")
````

````python
# ЗАДАЧА 3: Вопрос-ответ (Question Answering)
qa_pipeline = pipeline("question-answering", model="Qwen/Qwen2.5-0.5B-Instruct")

context = """
Москва — столица Российской Федерации, крупнейший город страны.
Основана в 1147 году князем Юрием Долгоруким.
Население составляет около 13 миллионов человек.
Москва является одним из крупнейших финансовых центров мира.
"""

question = "Кто основал Москву?"

result = qa_pipeline(question=question, context=context)
print(f"\nВопрос: {question}")
print(f"Ответ: {result['answer']}")
print(f"Уверенность: {result['score']:.2%}")
````

````python
# ЗАДАЧА 4: Суммаризация текста
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

text = """
Искусственный интеллект — это интеллект, демонстрируемый машинами, 
в отличие от естественного интеллекта, проявляемого людьми и животными. 
Ведущие учебники по ИИ определяют эту область как изучение интеллектуальных агентов: 
любых устройств, которые воспринимают окружающую среду и предпринимают действия, 
максимизирующие шансы на успешное достижение поставленных целей. 
В обиходе термин искусственный интеллект часто используется для описания машин, 
которые имитируют когнитивные функции, ассоциируемые с человеческим разумом, 
такие как обучение и решение проблем.
"""

input_text = "summarize: " + text

# Токенизация
inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

# Генерация
outputs = model.generate(
    inputs['input_ids'],
    max_length=100,
    min_length=25,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True,
    pad_token_id=tokenizer.eos_token_id
)

# Декодирование
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\nОригинальный текст:")
print(text.strip())
print("\nСуммаризация:")
print(summary)
````

### 4.3. Ручная загрузка модели и токенизатора

Для большего контроля загружаем модель и токенизатор отдельно.

````python
# Загрузка токенизатора и модели Qwen-2.5
model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print(f"Модель загружена: {model_name}")
print(f"Размер словаря: {tokenizer.vocab_size}")
````

````python
# Токенизация текста
text = "Машинное обучение — это"

# Кодирование
inputs = tokenizer(text, return_tensors="pt")
print(f"Входные токены: {inputs['input_ids'].tolist()[0]}")
print(f"Текст → Токены: '{text}' → {inputs['input_ids'].shape}")

# Декодирование (обратно)
decoded = tokenizer.decode(inputs['input_ids'][0])
print(f"Токены → Текст: {decoded}")
````

### 4.4. Генерация текста с параметрами

Исследуем влияние параметров на качество генерации.

````python
# Функция для генерации с параметрами
def generate_text(prompt, temperature=1.0, top_k=50, top_p=1.0, max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt")
    
    outputs = model.generate(
        inputs['input_ids'],
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Тестовый промпт
prompt = "Будущее искусственного интеллекта"
print(f"Промпт: '{prompt}'\n")
````

````python
# Эксперимент 1: Разная температура
print("=== Влияние температуры (temperature) ===")
for temp in [0.1, 0.7, 1.0, 1.5]:
    result = generate_text(prompt, temperature=temp)
    print(f"\nTemperature = {temp}:")
    print(f"  {result}")
````

````python
# Эксперимент 2: Top-K sampling
print("\n=== Влияние top_k ===")
for k in [1, 10, 50, 100]:
    result = generate_text(prompt, temperature=0.8, top_k=k)
    print(f"\nTop-K = {k}:")
    print(f"  {result}")
````

````python
# Эксперимент 3: Nucleus sampling (top_p)
print("\n=== Влияние top_p (nucleus sampling) ===")
for p in [0.5, 0.8, 0.95, 1.0]:
    result = generate_text(prompt, temperature=0.8, top_k=50, top_p=p)
    print(f"\nTop-P = {p}:")
    print(f"  {result}")
````

### 4.5. Сравнение моделей для классификации

Сравним несколько моделей для одной задачи.

````python
# Модели для классификации тональности
models = [
    "distilbert-base-uncased-finetuned-sst-2-english",
    "cardiffnlp/twitter-roberta-base-sentiment",
    "finiteautomata/bertweet-base-sentiment-analysis"
]

test_texts = [
    "Абсолютно фантастически! Лучшая покупка в моей жизни!",
    "Ужасное качество, полная трата денег.",
    "В целом нормально, ничего особенного."
]

print("=== Сравнение моделей классификации ===\n")

for model_name in models:
    print(f"\nМодель: {model_name}")
    classifier = pipeline("sentiment-analysis", model=model_name)
    
    for text in test_texts:
        result = classifier(text[:100])[0]  # Обрезаем для BERTweet
        print(f"  Текст: '{text[:50]}...'")
        print(f"    → {result['label']} ({result['score']:.2%})")
    print()
````

### 4.6. Индивидуальное задание: Модели для русского языка

Теперь протестируем модели, обученные на русском языке.

````python
# ВАРИАНТ A: ruBERT (классификация тональности)
print("\n=== ruBERT: Классификация тональности ===")

try:
    ru_classifier = pipeline(
        "sentiment-analysis",
        model="blanchefort/rubert-base-cased-sentiment"
    )
    
    ru_texts = [
        "Отличный продукт! Очень доволен покупкой.",
        "Ужасное качество, деньги на ветер.",
        "Нормально, ничего особенного."
    ]
    
    for text in ru_texts:
        result = ru_classifier(text)[0]
        print(f"\nТекст: {text}")
        print(f"  → {result['label']} (уверенность: {result['score']:.2%})")
        
except Exception as e:
    print(f"Ошибка загрузки ruBERT: {e}")
````

````python
# ВАРИАНТ B: Qwen2.5-3B-Instruct (вопрос-ответ)
print("\n=== Qwen2.5-3B-Instruct: Вопрос-ответ ===")

try:
    ru_qa = pipeline(
        "question-answering",
        model="Qwen/Qwen2.5-3B-Instruct"
    )
    
    ru_context = """
Санкт-Петербург — второй по численности населения город России.
Основан императором Петром I в 1703 году.
Был столицей Российской империи более 200 лет.
Население составляет около 5,6 миллионов человек.
"""
    
    ru_question = "Когда был основан Санкт-Петербург?"
    
    result = ru_qa(question=ru_question, context=ru_context)
    print(f"Вопрос: {ru_question}")
    print(f"Ответ: {result['answer']}")
    print(f"Уверенность: {result['score']:.2%}")
    
except Exception as e:
    print(f"Ошибка загрузки Qwen2.5-3B-Instruct: {e}")
````
